```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_fwf('/content/sales.txt',colspecs='infer',header=None)
#dataset from https://econometrics.com/intro/sales.txt
```

```python
print(df.shape) #dimensions of df
```

```python
df.head() #top 5 rows
```

```python
df.columns = ['Advertising', 'Sales'] #Renaming the columns of the dataset
df.head()
```

```python
df.info() #df summary
```

```python
df.describe()
```

```python
#independent and dependent variables
X = df['Advertising'].values
y=df['Sales'].values
```

```python
#Visual exploratory analysis
#Visualize relationship between X and y using scatter plot.
plt.scatter(X,y,color='blue',label='Scatter Plot')
plt.title('Relationship between Sales and Advertising')
plt.xlabel('Advertising')
plt.ylabel('Sales')
plt.legend(loc=4)
plt.show()
```

```python
print(X.shape)
print(y.shape)
```

```python
X=X.reshape(-1,1)
y=y.reshape(-1,1)
#Printing the dimensions after reshaping
print(X.shape)
print(y.shape)
```

```python
#Split X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,random_state=42)
```

```python
#print the dimensions of X_train, X_test, y_train, y_test
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Fit linear model
#Instantiate the linear regression object lm
from sklearn.linear_model import LinearRegression
lm = LinearRegression()

#train the model using training data sets
lm.fit(X_train,y_train)

#predict on the test data
y_pred = lm.predict(X_test)
```

```python
# y = 1.60509347 * x - 11.16003616
#Compute model slope and intercept

a= lm.coef_
b=lm.intercept_
print("Estimated model slope, a:", a)
print("Estimated model intercept, b:", b)
```

```python
#So, our fitted regression line is
# y = 1.60509347 * x - 12.45519409
```

```python
lm.predict(X) [0:5]
```

```python
#Without [0:5], we will get advertising values of whole sales dataset
lm.predict(X)
```

```python
#To make an individual prediction using Linear regression model
print(lm.predict([[24]]))
```

```python
#Calculate and print root mean square error (RMSE)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("Root Mean Square Error: {:.4f}".format(rmse))
```

```python
#r2 score
from sklearn.metrics import r2_score
print("R2 score: {:.4f}".format(r2_score(y_test, y_pred)))
```

```python
#plot of refression line

plt.scatter(X,y,color='blue',label = 'Scatter Plot')
plt.plot(X_test,y_pred,color='k',linewidth=3,label='Regression Line')
plt.title('Relationship between Sales and Advertising')
plt.xlabel('Advertising')
plt.ylabel('Sales')
plt.legend(loc=4)
plt.show()
```

```python
#check overfitting or underfitting of the data
print("Training set score : {:.4f}".format(lm.score(X_train, y_train)))
print("Test set score : {:.4f}".format(lm.score(X_test, y_test)))
```
